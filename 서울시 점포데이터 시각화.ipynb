```python
df2=pd.read_csv("C:/Users/82107/OneDrive/문서/2024-1/데베관1차/마포구 음식 시계열 점포수개폐업률.csv")
```

```python
df2
```

```python
# 날짜 문자열로 변환
df2['기준_년분기_코드'] = df2['기준_년분기_코드'].astype(str)

# 자치구별로 그룹화된 데이터 프레임 생성
grouped = df2.groupby(['자치구_코드_명', '업종_코드_명'])

# 그래프를 그릴 업종 리스트
target_industries = ['호프-간이주점', '한식음식점', '치킨전문점', '커피-음료']

# 그래프 그리기
plt.figure(figsize=(9, 7))

for industry in target_industries:
    for name, group in grouped:
        if industry in group['업종_코드_명'].unique():
            plt.plot(group['기준_년분기_코드'], group[group['업종_코드_명'] == industry]['평균폐업률'], marker='o', label=f'{industry}')
        
plt.xlabel('연분기')
plt.ylabel('폐업률(%)')
plt.title('업종별 폐업률 추세')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df3['연도분기'] = df3['연도분기'].astype(str)

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 점포수 막대 그래프
ax1.bar(df3['연도분기'], df3['점포수평균'], color='skyblue', label='평균 점포수')
ax1.set_ylabel('평균 점포수')

# ax1 오른쪽에 개업률, 폐업률 꺾은선 그래프 추가
ax2 = ax1.twinx()
ax2.plot(df3['연도분기'], df3['평균개업율'], marker='o', color='green', linestyle='-', label='평균 개업율')
ax2.plot(df3['연도분기'], df3['평균폐업률'], marker='o', color='red', linestyle='-', label='평균 폐업률')
ax2.set_ylabel('개업률 / 폐업률')

# 그래프 제목과 축 레이블 설정
plt.title('커피-음료 년도별 평균 점포수, 개업율, 폐업률')
ax1.set_xlabel('년도')

# 범례 표시
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='best')

# 그래프 표시
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_2023 = df3[df3['기준년도'] == 2023]

# 년도별로 그룹화하여 점포수평균, 개업율, 폐업률이 가장 높은 업종 5개 선택
top_avg_stores = df_2023.nlargest(5, '점포수평균')
top_opening_rate = df_2023.nlargest(5, '평균개업율')
top_closing_rate = df_2023.nlargest(5, '평균폐업률')

# 그래프 그리기
plt.figure(figsize=(8, 13))

# 점포수 평균 그래프
plt.subplot(3, 1, 1)
for year, group in top_avg_stores.groupby('기준년도'):
    plt.barh(group['업종_코드_명'], group['점포수평균'], label=year)
plt.title('Top5 점포수 업종')
plt.xlabel('점포수평균')
plt.ylabel('업종')
plt.legend()

# 개업율 그래프
plt.subplot(3, 1, 2)
for year, group in top_opening_rate.groupby('기준년도'):
    plt.barh(group['업종_코드_명'], group['평균개업율'], label=year)
plt.title('Top 5 개업율 업종')
plt.xlabel('개업률')
plt.ylabel('업종')
plt.legend()

# 폐업률 그래프
plt.subplot(3, 1, 3)
for year, group in top_closing_rate.groupby('기준년도'):
    plt.barh(group['업종_코드_명'], group['평균폐업률'], label=year)
plt.title('Top 5 폐업률 업종')
plt.xlabel('폐업률')
plt.ylabel('업종')
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
df4=pd.read_csv('C:/Users/82107/OneDrive/문서/2024-1/데베관1차/년도별직장인구데이터.csv')
df4
```

```python
# 각 년도별로 직장인구수가 가장 높은 5개 지역 선택
top5_resident_population = df4.groupby('기준년도').apply(lambda x: x.nlargest(5, '직장인구_수_평균')).reset_index(drop=True)

# 그래프 그리기
for year, data in top5_resident_population.groupby('기준년도'):
    plt.figure(figsize=(10, 6))
    plt.bar(data['자치구_코드_명'], data['직장인구_수_평균'])
    plt.title(f'Top 직장인구수 지역 {year}')
    plt.xlabel('지역')
    plt.ylabel('평균직장인구수')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
